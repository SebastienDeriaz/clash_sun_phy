```python
import numpy as np
import pickle
import matplotlib.pyplot as plt

with open('data.pkl', 'rb') as f:
    data = pickle.load(f)

#data
print(list(data))
```

```python
signals = {
    1 : ['phr_encoder_ready_i', 'phr_encoder_valid_o'],
    2 : ['phr_interleaver_ready_i', 'phr_interleaver_valid_o'],
    3 : ['phr_ofdm_ready_i', 'phr_ofdm_valid_o', 'ofdm_state'],
    4 : ['valid_o (actual)']
}

N = np.max(list(signals.keys()))

plt.figure(figsize=(18,8))
for k, lst in signals.items():
    plt.subplot(N*100 + 10 + k)
    plt.grid()
    for i, n in enumerate(lst):
        arr = np.asarray(data[n], dtype=int)
        plt.step(arr - i * 0.03, '.-', label=n)

    plt.xlim(0, 350)
    plt.legend()
plt.tight_layout()
plt.show()
```
